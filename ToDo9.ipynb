O segundo (superior à direita) são os mais **críticos** com mais alta frequencia de serem citados e também com maior impacito na decisão. Os do terceito quadrante (inferior à esquerda) são aqueles pouco citados e que afetam menos a decisão. Já o do quarto quadrante são aqueles que aparecem com muita frequência mas que pouco impactam na decisão.
